## Checking Outlier

```python
Almost = ['Sale_amount','Profit','Retail_price']
```

```python
plt.figure(figsize=(20,30))
for i in range(len(Almost)):
    plt.subplot(7,2, i+1)
    sns.boxplot(x = df[Almost[i]])
```

```python
#fill data that have small % missing values with median (because there's outlier)

df['Sale_amount'] = df['Sale_amount'].fillna(df['Sale_amount'].median())
df['Profit'] = df['Profit'].fillna(df['Profit'].median())
df['Retail_price'] = df['Retail_price'].fillna(df['Retail_price'].median())
```

```python
df.info()
```
